```python
%%writefile dashboard.py
import streamlit as st
import pandas as pd
import altair as alt
```

```python
import pandas as pd
data = pd.read_csv("data_bersih.csv")
data
```

```python
total_balita = len(data)
avg_birthweight = round(data['I05A'].mean(),2)
avg_birthlength = round(data['I07'].mean(),2)
avg_gest_age = round(data['I04'].mean(),2)
```

```python
import plotly.express as px
# a. Tren Rata-Rata Berat Lahir Berdasarkan Usia Kehamilan
trends_birthweight = data.groupby('I04')['I05A'].mean().reset_index()
fig_trend = px.line(trends_birthweight, x='I04', y='I05A',
                    title='Tren Rata-Rata Berat Lahir Berdasarkan Usia Kehamilan',
                    labels={'I04':'Usia Kehamilan (minggu)','I05A':'Rata-rata Berat Lahir (gr)'})
```

```python
# b. Distribusi Panjang Lahir Balita
fig_length_dist = px.histogram(data, x='I07', nbins=30, marginal='box',
                               title='Distribusi Panjang Lahir Balita',
                               labels={'I07':'Panjang Lahir (cm)'})
```

```python
# c. Perbandingan Berat dan Panjang Lahir Berdasarkan Jenis Kelamin
fig_gender_weight = px.box(data, x='B4K4', y='I05A', points='all', color='B4K4',
                           title='Perbandingan Berat Lahir Berdasarkan Jenis Kelamin',
                           labels={'B4K4':'Jenis Kelamin','I05A':'Berat Lahir (gr)'})
fig_gender_length = px.box(data, x='B4K4', y='I07', points='all', color='B4K4',
                           title='Perbandingan Panjang Lahir Berdasarkan Jenis Kelamin',
                           labels={'B4K4':'Jenis Kelamin','I07':'Panjang Lahir (cm)'})
```

```python
# d. Hubungan Panjang dan Berat Lahir (Scatter Plot)
fig_scatter = px.scatter(data, x='I07', y='I05A', color='B4K4',
                         title='Hubungan Panjang dan Berat Lahir',
                         labels={'I07':'Panjang Lahir (cm)','I05A':'Berat Lahir (gr)','B4K4':'Jenis Kelamin'})
```

```python
# e. Distribusi Balita Berdasarkan Provinsi
prov_summary = data.groupby('B1R1').agg(
    count=('B4K4','count'),
    avg_weight=('I05A','mean'),
    avg_length=('I07','mean')
).reset_index()
fig_prov = px.bar(prov_summary, x='B1R1', y='count',
                  hover_data=['avg_weight','avg_length'],
                  title='Distribusi Balita Berdasarkan Provinsi',
                  labels={'B1R1':'Provinsi','count':'Jumlah Balita'})
```
